### Part 2: Visit Links, Gather University Data, and Perform Initial Clean ###

**Code for part two is found in the executables folder under "find_universities.py". The code runs in the following steps:**
1. For each faculty member in the data dictionary, first try to visit the "linked page" to gather the university data
2. Pull out all text on the page tagged with "\<p\>".
3. For those tags, if the tag contains words associated with  education, pull those tags out
4. If no education data is found, repeat steps 2 and 3 with the directory pages
5. Clean the education data if it was found in list form (not actually written within a paragraph). Replace all education pertaining to doctorate degree with "PhD". Once the university data is clean, return it as a tuple of (degree, school, year)
6. Write the partially cleaned education data to the json file specified

### We observed some similarities across these two departments:
1. Both SSD and Harris **love hiring PhDs from Uchicago and Ivy league**
2. **Ivy league** dominates **>35% of faculty pool**
3. Both divisions hire predominantly faculty who completed their PhD **from the US**, with some level of exceptions.

### While there are also some differences that are worth noting:
1. Harris hires **many that dont have a PhD degree**: This could be due to the nature of public policy. It is a more "applied" field compared to social science, and therefore many faculty doesnt have PhD but has a lot of industry experience.
2. Harris is more **international** when it comes to hiring: This could be due to Harris' preference for hiring faculty who have experience practicing policy solutions around the world

### regression analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
from sklearn.preprocessing import maxabs_scale
import numpy as np
```

```python
# Scale the ranking to the [-1, 1] range without breaking the sparsity
uni_rank['Normalized Ranking'] = uni_rank['Ranking'].apply(lambda x: 1/x)
uni_rank['Normalized Ranking'] = maxabs_scale(np.array(uni_rank['Normalized Ranking']).reshape(-1,1))

uni_rank2['Normalized Ranking'] = uni_rank2['Ranking'].apply(lambda x: 1/x)
uni_rank2['Normalized Ranking'] = maxabs_scale(np.array(uni_rank2['Normalized Ranking']).reshape(-1,1))
```

```python
sns.set_style("darkgrid")
sns.lmplot(data = uni_rank, 
           x = 'Normalized Ranking',
           y = 'Number of Faculty',
           );
plt.title('Number of hires by university ranking in Social Science Division', size = 12)
plt.xlabel('University Ranking Index', size = 12)
plt.ylabel('Number of Hires', size = 12);
plt.savefig('SSD-hires-by-ranking.png', dpi=800)
```

```python
sns.set_style("darkgrid")
sns.lmplot(data = uni_rank2, 
           x = 'Normalized Ranking',
           y = 'Number of Faculty',
           );
plt.title('Number of hires by university ranking in Harris Public Policy School', size = 15)
plt.xlabel('University Ranking Index', size = 12)
plt.ylabel('Number of Hires', size = 12);
plt.savefig('Harris-hires-by-ranking.png')
```

```python
# run OLS
y = uni_rank['Number of Faculty']
x = uni_rank['Normalized Ranking']
ols_model = sm.OLS(y, x)  
results = ols_model.fit() 
print(results.summary())
```

```python
# run OLS
y2 = uni_rank2['Number of Faculty']
x2 = uni_rank2['Normalized Ranking']
ols_model2 = sm.OLS(y2, x2)  
results2 = ols_model2.fit() 
print(results2.summary())
```

```python
# scatterplot
sns.relplot(data = uni_rank,
            x = 'Normalized Ranking',
            y = 'Number of Faculty',
            kind = 'scatter');

# plot regression line
plt.title('Social Science Division: Fitted Regression Line', size = 15)
plt.plot(uni_rank['Normalized Ranking'], results.fittedvalues);
plt.savefig('SSD-regression.png')
```

```python
# plot residuals
plt.scatter(uni_rank['Normalized Ranking'], results.resid)

# improve graph with horizontal line, title, etc.
plt.axhline(y = 0)
plt.title('Social Science Division:  Residual Plot', size = 15);
plt.savefig('SSD-residual.png')
```

```python
# scatterplot
sns.relplot(data = uni_rank2,
            x = 'Normalized Ranking',
            y = 'Number of Faculty',
            kind = 'scatter');

# plot regression line
plt.title('Harris Public Policy School: Fitted Regression Line', size = 15)
plt.plot(uni_rank2['Normalized Ranking'], results2.fittedvalues);
plt.savefig('Harris-regression.png')
```

```python
# plot residuals
plt.scatter(uni_rank2['Normalized Ranking'], results2.resid)

# improve graph with horizontal line, title, etc.
plt.axhline(y = 0)
plt.title('Harris Public Policy School: Residual Plot', size = 15);
plt.savefig('Harris-residual.png')
```
